```python
class MyStack():
    """A simple stack implementation."""
    def __init__(self):
        self.data = []
    def my_push(self, value = None):
        """Pushes a value onto the stack."""
        if value == None:
            return None
        else:
            self.data.append(value)
            return value
        
    def my_pop(self):
        """Pops a value off of the stack."""
        if len(self.data) == 0:
            return None
        else:
            popped_value = self.data[-1]
            self.data = self.data[:-1]
            return popped_value   
```

```python
class ListNode(value = None, next_node = None):
    """Create a new node with a value and a pointer to a single child node."""
    def __init__(self):
        self.value = value
        self.next_node = next_node

class SinglyLinkedList(ListNode, value = None):
    """Create a singly linked list object with a head with the specified defaults, and the methods below."""
    def __init__(self):
        self.head = ListNode(value, None)
    
    def insert_node(self, value):
        """Insert the node at the specified value"""
        new_node = ListNode(value, self.head)
        self.head = new_node
        return new_node
    
    def delete_node(self, value):
        """Find the first instance of the given value in the list, and delete the corresponding node."""
        current_node = self.head
        while current_node != None:
            if current_node.value != value:
                current_node = current_node.next_node
            else:
                return current_node.value
        return None
    
    def delete_position(self, position):
        """Delete a node at the given position in the list if it exists."""
        previous_node = None
        current_node = self.head
        while position > -1:
            if current_node.next_node == None:
                return False
            elif position == 0:
                if previous_node == None:
                    self.head = self.head.next_node
                else:
                    previous_node.next_node = current_node.next_node
                return True
            position -= position
            previous_node = current_node
            curent_node = current_node.next_node
                
    def pop(self):
        """Pop a node off the head of the list."""
        self.head = self.head.next_node
    
    def push(self, value):
        """Push a new node on the head of the list."""
        new_node = ListNode(value, self.head)
        self.head = new_node
    
    
```
